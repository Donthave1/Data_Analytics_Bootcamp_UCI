```python
# Dependencies and Setup
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
clinical_data = pd.merge(drug_data, trial_data, on=["Mouse ID"]).dropna(how="any")

# Display the data table for preview
clinical_data.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_drug_timepoint = clinical_data.groupby(["Drug", "Timepoint"])
tumor_vol_avg = group_drug_timepoint["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
tumor_response = pd.DataFrame({"Tumor Volume (mm3)": tumor_vol_avg}).reset_index()
# Preview DataFrame
tumor_response.head(11)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = group_drug_timepoint["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
tumor_standard_error = pd.DataFrame({"Tumor Volume Standard Error": tumor_sem}).reset_index()
# Preview DataFrame
tumor_standard_error.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
vol_tp_table = tumor_vol_avg.unstack(level=0)
# Preview that Reformatting worked
vol_tp_table.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = vol_tp_table.index

capo_resp = plt.errorbar(x_axis, vol_tp_table["Capomulin"], yerr=vol_tp_table["Capomulin"].sem(), fmt="ro-")
infu_resp = plt.errorbar(x_axis, vol_tp_table["Infubinol"], yerr=vol_tp_table["Infubinol"].sem(), fmt="b^-")
keta_resp = plt.errorbar(x_axis, vol_tp_table["Ketapril"], yerr=vol_tp_table["Ketapril"].sem(), fmt="gs-")
plac_resp = plt.errorbar(x_axis, vol_tp_table["Placebo"], yerr=vol_tp_table["Placebo"].sem(), fmt="kd-")

# Save the Figure
plt.title("Tumor Response to Treatment", fontsize=12, fontweight="bold")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
plt.tight_layout()
plt.legend()

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site_avg = group_drug_timepoint["Metastatic Sites"].mean()
# Convert to DataFrame
metastatic_table = pd.DataFrame({"Metastatic Sites": met_site_avg}).reset_index()
# Preview DataFrame
metastatic_table.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem = group_drug_timepoint["Metastatic Sites"].sem()
# Convert to DataFrame
met_standard_error = pd.DataFrame({"Metastatic Sites Standard Error": met_sem}).reset_index()
# Preview DataFrame
met_standard_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_reformat = met_site_avg.unstack(level=0)
# Preview that Reformatting worked
met_reformat.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = met_reformat.index

capo_met = plt.errorbar(x_axis, met_reformat["Capomulin"], yerr=met_reformat["Capomulin"].sem(), fmt="ro:")
infu_met = plt.errorbar(x_axis, met_reformat["Infubinol"], yerr=met_reformat["Infubinol"].sem(), fmt="b^:")
keta_met = plt.errorbar(x_axis, met_reformat["Ketapril"], yerr=met_reformat["Ketapril"].sem(), fmt="gs:")
plac_met = plt.errorbar(x_axis, met_reformat["Placebo"], yerr=met_reformat["Placebo"].sem(), fmt="kd:")
# Save the Figure
plt.title("Metastatic Spread During Treatment", fontsize=12, fontweight="bold")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.grid()
plt.tight_layout()
plt.legend()

# Show the Figure
plt.show()
```
