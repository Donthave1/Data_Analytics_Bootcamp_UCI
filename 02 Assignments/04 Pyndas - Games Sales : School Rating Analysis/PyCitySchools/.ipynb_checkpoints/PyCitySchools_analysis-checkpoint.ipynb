# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].drop_duplicates())
#Calculate the total number of students
total_students = school_data_complete["student_name"].count()
#Calculate the total budget
total_budgets = school_data_complete["budget"].drop_duplicates().sum()
#Calculate the average math score
math_avg = school_data_complete["math_score"].mean()
#Calculate the average reading score
reading_avg = school_data_complete["reading_score"].mean()
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_rate = (math_avg + reading_avg) / 2
#Calculate the percentage of students with a passing math score (70 or greater)
math_passing_pct = len(school_data_complete[school_data_complete["math_score"] >= 70]) / total_students *100
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_passing_pct = len(school_data_complete[school_data_complete["reading_score"] >= 70]) / total_students *100
#Create Data Frame to hold all calculated values
district_summary = pd.DataFrame({
    "Total Schools": total_schools, 
    "Total Students": total_students, 
    "Total Budget": total_budgets, 
    "Average Math Score": math_avg, 
    "Average Reading Score": reading_avg, 
    "% Passing Math": [math_passing_pct], 
    "% Passing Reading": [reading_passing_pct], 
    "% Overall Passing Rate": [overall_rate]
})
#Reformat
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.6f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.6f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.6f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.6f}%".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:.6f}%".format)

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Create School Index table
school_table = school_data_complete.set_index("school_name").groupby(["school_name"])
#School Type
school_type = school_data.set_index("school_name")["type"]
#Total Students
school_students = school_table["Student ID"].count()
#Total School Budget
school_budget = school_table["budget"].mean()
#Per Student Budget
per_student_budget = (school_budget / school_students)
#Average Math Score
school_math_avg = school_table["math_score"].mean()
#Average Reading Score
school_reading_avg = school_table["reading_score"].mean()
#% Passing Math
school_math_pass = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/school_students * 100
#% Passing Reading
school_reading_pass = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/school_students * 100
#Overall Passing Rate (Average of the above two)
school_overall_pass = (school_math_pass + school_reading_pass) / 2

#Create a dataframe to hold the above results

school_summary = pd.DataFrame({
    "School Type": school_type, 
    "Total Students": school_students, 
    "Total School Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_math_avg,
    "Average Reading Score": school_reading_avg,
    "% Passing Math": school_math_pass,
    "% Passing Reading": school_reading_pass,
    "Overall Passing Rate": school_overall_pass
})

#Reformat
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.6f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.6f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:,.6f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.6f}%".format)
school_summary["Overall Passing Rate"] = school_summary["Overall Passing Rate"].map("{:.6f}%".format)

school_summary.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Filter each grade of math scores by each school
#9th grades
freshman_math = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name")["math_score"].mean()
#10th grades
sophomore_math = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name")["math_score"].mean()
#11th grades
junior_math = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name")["math_score"].mean()
#12th grades
senior_math = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name")["math_score"].mean()
#Create DataFrame to hold values
math_scores_by_grades = pd.DataFrame({
        "9th Math Score Avg": freshman_math,
        "10th Math Score Avg": sophomore_math,
        "11th Math Score Avg": junior_math,
        "12th Math Score Avg": senior_math
})

math_scores_by_grades 
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins
bins = [0, 584.99, 614.99, 644.99, 5000]
# Create labels
spending_range = ["< $585", "$585 - 615", "$615 - 645", "$645 - 675"]

# Cut the bins/label into DataFrame
new_speding_df = school_data_complete
new_speding_df["Spending Range"] = pd.cut(new_speding_df["budget"]/new_speding_df["Student ID"], bins, labels = spending_range)
# Group a Spending table
spending_table = new_speding_df.groupby("Spending Range")
#Spending students headcounts
spending_student_count = spending_table["Student ID"].count()
#Average Math Score
spending_math_avg = new_speding_df["math_score"].mean()
#Average Reading Score
spending_reading_avg = new_speding_df["reading_score"].mean()
#% Passing Math
spending_math_pass = new_speding_df[new_speding_df["math_score"] >= 70].groupby("Spending Range")["Student ID"].count() / spending_student_count * 100
#% Passing Reading
spending_reading_pass = new_speding_df[new_speding_df["reading_score"] >= 70].groupby("Spending Range")["Student ID"].count() / spending_student_count * 100 
#Overall Passing Rate (Average of the above two)
speding_overall_pass = (spending_math_pass + spending_reading_pass) / 2
            
# Create DataFrame            
spending_by_school = pd.DataFrame({
    "Average Math Score": spending_math_avg,
    "Average Reading Score": spending_reading_avg,
    '% Passing Math': spending_math_pass,
    '% Passing Reading': spending_reading_pass,
    "Overall Passing Rate": speding_overall_pass        
})

# Reformat
spending_by_school["Average Math Score"] = spending_by_school["Average Math Score"].map("{:.6f}".format)
spending_by_school["Average Reading Score"] = spending_by_school["Average Reading Score"].map("{:.6f}".format)
spending_by_school["% Passing Math"] = spending_by_school["% Passing Math"].map("{:,.6f}%".format)
spending_by_school["% Passing Reading"] = spending_by_school["% Passing Reading"].map("{:.6f}%".format)
spending_by_school["Overall Passing Rate"] = spending_by_school["Overall Passing Rate"].map("{:.6f}%".format)

spending_by_school
```

## Scores by School Size
