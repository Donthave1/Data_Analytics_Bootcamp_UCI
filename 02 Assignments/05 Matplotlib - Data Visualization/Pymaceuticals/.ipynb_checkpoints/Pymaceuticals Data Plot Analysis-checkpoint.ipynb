```python
# Dependencies and Setup
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
clinical_data = pd.merge(drug_data, trial_data, on=["Mouse ID"]).dropna(how="any")

# Display the data table for preview
clinical_data.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_drug_timepoint = clinical_data.groupby(["Drug", "Timepoint"])
tumor_vol_avg = group_drug_timepoint["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
tumor_response = pd.DataFrame({"Tumor Volume (mm3)": tumor_vol_avg}).reset_index()
# Preview DataFrame
tumor_response.head(11)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = group_drug_timepoint["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
tumor_standard_error = pd.DataFrame({"Tumor Volume Standard Error": tumor_sem}).reset_index()
# Preview DataFrame
tumor_standard_error.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
vol_tp_table = tumor_vol_avg.unstack(level=0)
# Preview that Reformatting worked
vol_tp_table.head()
```

```python
# Generate the Plot (with Error Bars)
vol_x_axis = vol_tp_table.index

capo_resp = plt.errorbar(vol_x_axis, vol_tp_table["Capomulin"], yerr=vol_tp_table["Capomulin"].sem(), fmt="ro-")
infu_resp = plt.errorbar(vol_x_axis, vol_tp_table["Infubinol"], yerr=vol_tp_table["Infubinol"].sem(), fmt="b^-")
keta_resp = plt.errorbar(vol_x_axis, vol_tp_table["Ketapril"], yerr=vol_tp_table["Ketapril"].sem(), fmt="gs-")
plac_resp = plt.errorbar(vol_x_axis, vol_tp_table["Placebo"], yerr=vol_tp_table["Placebo"].sem(), fmt="kd-")

# Save the Figure
plt.title("Tumor Response to Treatment", fontsize=12, fontweight="bold")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
plt.tight_layout()
plt.legend()

# Save Figure
plt.savefig("Graph Images/Tumor Response to Treatment ErrorBar Graph.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site_avg = group_drug_timepoint["Metastatic Sites"].mean()
# Convert to DataFrame
metastatic_table = pd.DataFrame({"Metastatic Sites": met_site_avg}).reset_index()
# Preview DataFrame
metastatic_table.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem = group_drug_timepoint["Metastatic Sites"].sem()
# Convert to DataFrame
met_standard_error = pd.DataFrame({"Metastatic Sites Standard Error": met_sem}).reset_index()
# Preview DataFrame
met_standard_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_reformat = met_site_avg.unstack(level=0)
# Preview that Reformatting worked
met_reformat.head()
```

```python
# Generate the Plot (with Error Bars)
met_x_axis = met_reformat.index

capo_met = plt.errorbar(met_x_axis, met_reformat["Capomulin"], yerr=met_reformat["Capomulin"].sem(), fmt="ro:")
infu_met = plt.errorbar(met_x_axis, met_reformat["Infubinol"], yerr=met_reformat["Infubinol"].sem(), fmt="b^:")
keta_met = plt.errorbar(met_x_axis, met_reformat["Ketapril"], yerr=met_reformat["Ketapril"].sem(), fmt="gs:")
plac_met = plt.errorbar(met_x_axis, met_reformat["Placebo"], yerr=met_reformat["Placebo"].sem(), fmt="kd:")
# Save the Figure
plt.title("Metastatic Spread During Treatment", fontsize=12, fontweight="bold")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.grid()
plt.tight_layout()
plt.legend()

# Save Figure
plt.savefig("Graph Images/Metastatic Spread During Treatment ErrorBar Graph.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = group_drug_timepoint["Mouse ID"].nunique()
# Convert to DataFrame
mouse_df = pd.DataFrame({"Mouse Count": mouse_count}).reset_index()
# Preview DataFrame
mouse_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_reformat = mouse_count.unstack(level=0)
# Preview the Data Frame
mouse_reformat.head()
```

```python
# Generate the Plot (Accounting for percentages)
mice_x_axis = mouse_reformat.index

capo_surv = plt.plot(mice_x_axis, mouse_reformat["Capomulin"], color="r", marker="o", linestyle="-.")
infu_surv = plt.plot(mice_x_axis, mouse_reformat["Infubinol"], color="b", marker="^", linestyle="-.")
keta_surv = plt.plot(mice_x_axis, mouse_reformat["Ketapril"], color="g", marker="s", linestyle="-.")
plac_surv = plt.plot(mice_x_axis, mouse_reformat["Placebo"], color="k", marker="d", linestyle="-.")

# Save the Figure
plt.title("Survival During Treatment", fontsize=12, fontweight="bold")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()
plt.tight_layout()
plt.legend()

# Save Figure
plt.savefig("Graph Images/Survival During Treatment ErrorBar Graph.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_begin_vol = vol_tp_table.iloc[0,:]
tumor_end_vol = vol_tp_table.iloc[-1,:]
tumor_percent_change = (tumor_end_vol / tumor_begin_vol) * 100 - 100

# Display the data to confirm
tumor_percent_change
```

```python
# Create a table just for the four drugs used in graph
test_drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
test_object = vol_tp_table[test_drugs]

# Calculate the percent changes for each drug
bar_begin_vol = test_object.iloc[0,:]
bar_end_vol = test_object.iloc[-1,:]
bar_percent_change = (bar_end_vol / bar_begin_vol) * 100 - 100

# Plot the grpah
vol_x_axis = [x for x in range(len(bar_percent_change))]

plt.bar(vol_x_axis, 
        bar_percent_change, 
        width = 1, 
        edgecolor="k",
        linewidth=2,
        color = ["r" if bar_percent_change[x] > 0 else "g" for x in np.arange(len(vol_x_axis))]
       )

# Use functions to label the percentages of changes
x_bar = 0
for x in bar_percent_change:
    if x < 0:
        label_position = -5
    else:
        label_position = 2
    plt.text(x_bar, label_position, str(round(x)) + "%", ha = "center", color = "w", fontweight="bold")
    x_bar += 1


# Configurate the Graph
plt.title("Tumor Change Over 45 Days Treatment", fontsize=12, fontweight="bold")
plt.ylabel("% Tumor Volume Change")
tick_locations = [x+0.5 for x in vol_x_axis]
plt.xticks(tick_locations, ["Capomulin", "Infubinol", "Ketapril", "Placebo"])
plt.xlim(-0.8, 3.8)
plt.ylim(-30, 70)
plt.grid()
plt.hlines(y=0, xmin=-50, xmax=max(vol_x_axis)+50, color="black")

# Save Figure
plt.savefig("Graph Images/Tumor Change Over 45 Days Treatment Bar Graph.png")

# Show the Figure
plt.show()
```

# Observation

* Based on "Tumor Response to Treatment" chart, we can clearly see that Capomulin has great effect of reduce tumor growth in test subject, while other comparison drug: Infubinol, Ketapril, and Placebo did not have much effect on restraining tumor growth.

* After examing Survival During Treatment graph, drug named Infubinol display a greater side effect that cause a greater death rate (lowest survive rate). Drug Ketapril and Placebo have similar survival rate where Placebo has better rate at initial stage but fall off quickly. On the other hand Ketapril had lowest survival rate at begining that may show test subject had hardest time intake the treatment, but for those who survived tend to last longer than object treated with Infubinol or Ketapril. Lastly, Capomulin demostrated the best survival rate of test subject, which echo our first observation. 

* In conclusion, Tumor Chnage Over 45 Days Treatment graph show case an overall tumor percent change in the past 45 days where each drug had successfully reduced tumor size or the drug is ineffective. Red bar represent growth of the tumor and green bar represent reduction of the tumor in test subjects.
