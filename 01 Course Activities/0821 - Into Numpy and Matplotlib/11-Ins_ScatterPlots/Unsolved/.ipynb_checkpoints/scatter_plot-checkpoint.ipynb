```python
# Import Dependencies
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_limit = 100
x_axis = np.arange(0, x_limit, 1)
data = [random.random() for value in x_axis]
```

```python
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black", 
           s=x_axis, alpha=0.75)
```

```python
plt.xlim(0, 100)
```

```python
plt.ylim(0, 1)
```
