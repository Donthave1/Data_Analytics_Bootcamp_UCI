```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(1, 13, 1)
x_axis
```

```python
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(x_axis, points)
plt.show()
```

```python
points_C = [(x-32)*0.56 for x in points]
points_C
```

```python
plt.plot(x_axis, points)
plt.plot(x_axis, points_C)
plt.show()
```

```python
F_handle, = plt.plot(x_axis, points, marker='+', color='blue', linewidth=1, label='Fahrenheit')
C_handle, = plt.plot(x_axis, points_C, marker='s', color='red', linewidth=1, label='Celcius')
```

```python
plt.legend(handles=[F_handle, C_handle], loc="best")
```

```python
plt.xlabel("Month")
plt.ylabel("Degrees")
```
