```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["Humans", "Smurfs", "Hobbits", "Ninjas"]
sizes = [220, 95, 80, 100]
color = ["red", "orange", "lightcoral", "lightskyblue"]
explode_value = (0, 0, 0, 0)
```

```python
plt.pie(sizes, explode=explode_value, labels=labels, colors=color, 
       autopct="%1.1f%%", shadow=True, startangle=140)
```

```python
plt.axis("equal")
```
