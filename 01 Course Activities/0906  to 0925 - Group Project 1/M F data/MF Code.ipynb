```python
# Dependencie
import pandas as pd
```

```python
# Load in file
marathon_2015 = "marathon_results_2015.csv"
```

```python
marathon_2015_pd = pd.read_csv(marathon_2015)
marathon_2015_pd.head()
```

```python
marathon_2015_pd.columns
```

```python
male_2015 = marathon_2015_pd.loc[marathon_2015_pd["M/F"] == "M", ['Unnamed: 0', 'Bib', 'Name', 'Age', 'M/F', 'City', 'State', 'Country',
       'Citizen', 'Unnamed: 9', '5K', '10K', '15K', '20K', 'Half', '25K',
       '30K', '35K', '40K', 'Pace', 'Proj Time', 'Official Time', 'Overall',
       'Gender', 'Division']]
male_2015.head
```

```python
male_2015.to_csv("male_marathon_2015.csv", index=False)
```
